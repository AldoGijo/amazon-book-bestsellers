```python
df=pd.get_dummies(df,columns=["Format"],drop_first=True,dtype=int)
```

```python
df.head()
```

```python
# Create a frequency feature: how many books does this author have in the top 500?
df['Author_Book_Count'] = df['Author'].map(df['Author'].value_counts())
df['Publisher_Book_Count'] = df['Publisher'].map(df['Publisher'].value_counts())
```

```python
df.info()
```

```python
df_cleaned=df.copy()
```

```python
current_year=2026
df_cleaned['Year_Since_Release']=current_year-df['Year Published'].dt.year
df_cleaned.drop(columns=['Title','Author','Sub-Genre','Primary_Sub_Genre','Publisher','Year Published'],inplace=True)
```

```python
df_cleaned.head()
```

```python
df_cleaned.info()
```

```python
df_cleaned.to_csv("Amazon_Book_Cleaned.csv",index=False)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
x=df_cleaned.drop(columns=['Weeks on List','Rank'])
y=df_cleaned['Weeks on List']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
model=RandomForestRegressor()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
print(y_pred)
```

```python
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
```

```python
print(f'mean_absolute_error:{mae}')
```

```python
print(f'r2_score:{r2}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Get feature importance values from the model
importances = model.feature_importances_

# 2. Create a clean dataframe mapping features to their importance score
feature_imp_df = pd.DataFrame({
    'Feature': x.columns,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

print("--- Feature Importance Breakdown ---")
print(feature_imp_df)

# 3. Plot the importances for clear visualization
plt.figure(figsize=(10, 6))
plt.barh(feature_imp_df['Feature'][::-1], feature_imp_df['Importance'][::-1], color='skyblue')
plt.xlabel('Importance Score')
plt.title('Which Features Drive a Book\'s Weeks on List?')
plt.tight_layout()
plt.show()
```

```python
import pickle

# Save the trained Random Forest model
with open('book_success_model.pkl', 'wb') as model_file:
    pickle.dump(model, model_file)

# Save your LabelEncoder (from earlier steps) so you can decode categories in Streamlit
with open('category_encoder.pkl', 'wb') as le_file:
    pickle.dump(le, le_file)

print("🎉 Model and encoders saved successfully as deployment files!")
```
